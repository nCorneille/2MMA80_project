# Getting Started with Jupyter Notebooks

This notebook will guide you through the basic functionality available in Jupyter notebooks. If you are already familiar with Jupyter notebooks, feel free to skip this tutorial.

A Jupyter notebook is a mixed collection of text, graphics and code. The setup is similar to Mathematica, but made to be language agnostic.

### Cells

A notebooks consits of cells such as this one. There are 3 types of cells: _Code_, _Markdown_ and _Raw_. Relevant to us are _Markdown_ and _Code_. If you click on this cell you will see in the toolbar at the top of the notebook a dropdown that says _Markdown_, this indicates that this is a cell of that type. The type of the cell can be changed by clicking on the type dropdown.

### Markdown

Markdown is a simple language to add formatting to plain text. You can see it in action by double clicking this cell, this will display the cell in _edit mode_ as plain text. Hitting `Ctrl` + `Enter` will execute the cell and turn it back into formated text.

You can do the typical formatting such as: _italic_, __bold__, `code`. You can make lists:
- item 1,
- item 2,

and tables:

| Nr. | Label    | Cost |
|-----|----------|------|
| 1   | Making   | 10   |
| 2   | a        | 5    |
| 3   | table    | 80   |

For an extensive overview of Markdown have a look at  [www.markdownguide.org](https://www.markdownguide.org/getting-started/).

Markdown cells also support $\LaTeX$. You can add math inline with \\$ ... \\$ or on its own line with \\$\\$ ... \\$\\$. For example:
$$
\int_a^b f(x) dx.
$$

### Keyboard shortcuts

Jupyter lab has many shortcuts that make editing and navigating notebooks easier, we list some common ones.

#### Shortcuts when in either _command mode_ (outside the cells) or _edit mode_ (inside a cell):

- `Shift` + `Enter` run selected cell or cells - if no cells below, insert a code cell below

- `Ctrl` + `S` save and checkpoint
- `Ctrl` + `F` find 

#### Shortcuts when in _command mode_ (outside the cells, no blinking cursor):

- `Enter` enter _edit mode_ in the active cell

- `A` insert a new cell above the active cell
- `B` insert a new cell below the active cell

- `M` make the active cell a Markdown cell
- `Y` make the active cell a code cell

- `Shift` + `Up Arrow` select the current cell and the cell above
- `Shift` + `Down Arrow` select the current cell and the cell below

- `X` cut the selected cell or cells
- `C` copy the selected cell or cells
- `V` paste the cell(s) which were copied or cut most recently

- `Shift + M` merge multiple selected cells into one cell

- `DD` (`D` twice) delete the active cell
- `00` (Zero twice) restart the kernel

- `Z` undo most recent command mode action

#### Shortcuts when in _edit mode_ (inside a cell with a blinking cursor):


- `Esc` enter _command mode_

- `Tab` code completion (or indent if at start of line)
- `Shift` + `Tab` tooltip help
- `Ctrl` + `Shift` + `-` split the active cell at the cursor

#### The usual commands for code editors:

- `Ctrl` + `]` indent
- `Ctrl` + `[` dedent

- `Ctrl` + `/` toggle comment

Plus the usual shortcuts for select all, cut, copy, paste, undo, etc.

### Code & Kernels

Executing code happens in _Code_ cells. The code in these cells is executed by a _kernel_. This is how Jupyter is language agnostic: there exist many kernels supporting many different programming languages.

We will be using the _Python_ kernel, specifically the python kernel from the conda environment we just made. You can tell which kernel this notebook is using at the top right, it should say `Python [conda env:mathnn]`, if not you can click on the kernel name and select it from the list.

A code cell is executed by selecting it and hitting `Ctrl` + `Enter`, try it below.

```python
def fib(n : int):
    if n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

n = 11
print(f"The {n}th Fibonacci number = {fib(n)}")
```

## Matplotlib

The environment we created includes the _Matplotlib_ package to make plots. The next cell contains a small example of its use.

## TQDM Progress Bars

The [TQDM](https://tqdm.github.io/) package provides a convenient way to automatically monitor a _for_-loop with a progress bar.

```python
from tqdm.notebook import trange, tqdm
from time import sleep

for i in trange(100): # will loop i from 0 to 99 but instead of range(100) we use TQDM's trange to get a progress bar that tracks the loop.
    # do something useful
    sleep(0.02)
```
